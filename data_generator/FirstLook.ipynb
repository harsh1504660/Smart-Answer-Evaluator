# Remvoing stop words 

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('history_p1_preprocessed.csv')
```

```python
data.head()
```

```python
data.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
data.head()
```

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
def remove_stopwords(text):
    stop_words = set(stopwords.words('english'))
    words = nltk.word_tokenize(text)
    filtered_text = [word for word in words if word.lower() not in stop_words]
    return ' '.join(filtered_text)
```

```python
text = "This is an example sentence to demonstrate removing stopwords."
filtered_text = remove_stopwords(text)
print(filtered_text)
```

```python
data['ideal answer'] = data['ideal answer'].apply(remove_stopwords)
data['students answer'] = data['students answer'].apply(remove_stopwords)
```

```python
data.head()
```

```python
data['ideal answer'] = data['ideal answer'].replace('\n','')
data['students answer'] = data['students answer'].replace('\n','')
```
